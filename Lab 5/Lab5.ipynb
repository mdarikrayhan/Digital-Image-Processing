```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('/Users/mdarikrayhan/Documents/GitHub/Digital-Image-Processing/Lab 5/Photo.jpg')

array = np.array([[255, 0, 1, 100, 25, 101], [10, 200, 70, 80, 120, 150], [
                 95, 30, 30, 81, 96, 77], [87, 89, 220, 250, 100, 10], [18, 7, 221, 21, 8, 15]])
temp = array.copy()

max_pixel = np.max(array)
#get the ceil of power of 2
max_pixel = (2**(np.ceil(np.log2(max_pixel))))-1

for i in range(0, 5):
    for j in range(0, 6):
        array[i][j] = max_pixel-array[i][j]

print(temp)
print(array)
```

```python
#negative of an image
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
temp = img_gray.copy()

max_pixel = np.max(img_gray)
#get the ceil of power of 2
max_pixel = (2**(np.ceil(np.log2(max_pixel))))-1 #L-1 = 255

height = img_gray.shape[0]
width = img_gray.shape[1]

for i in range(0, height):
    for j in range(0, width):
        img_gray[i][j] = max_pixel-img_gray[i][j]

plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(temp, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(img_gray, cmap='gray')
```

```python
# power law transformation
img_gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gray_125 = img_gray.copy()
img_gray_25 = img_gray.copy()
img_gray_50 = img_gray.copy()
img_gray25 = img_gray.copy()
img_gray50 = img_gray.copy()
img_gray100 = img_gray.copy()

height = img_gray.shape[0]
width = img_gray.shape[1]

for i in range(0, height):
    for j in range(0, width):
        img_gray_125[i][j] = np.power(img_gray[i][j], 0.125)
        img_gray_25[i][j] = np.power(img_gray[i][j], 0.25)
        img_gray_50[i][j] = np.power(img_gray[i][j], 0.5)
        img_gray25[i][j] = np.power(img_gray[i][j], 2.5)
        img_gray50[i][j] = np.power(img_gray[i][j], 5)
        img_gray100[i][j] = np.power(img_gray[i][j], 10)

plt.figure(figsize=(10, 10))

plt.subplot(2, 3, 1)
plt.imshow(img_gray_125, cmap='gray')
plt.title('gamma=0.125')
plt.subplot(2, 3, 2)
plt.imshow(img_gray_25, cmap='gray')
plt.title('gamma=0.25')
plt.subplot(2, 3, 3)
plt.imshow(img_gray_50, cmap='gray')
plt.title('gamma=0.5')
plt.subplot(2, 3, 4)
plt.imshow(img_gray25, cmap='gray')
plt.title('gamma=2.5')
plt.subplot(2, 3, 5)
plt.imshow(img_gray50, cmap='gray')
plt.title('gamma=5')
plt.subplot(2, 3, 6)
plt.imshow(img_gray100, cmap='gray')
plt.title('gamma=10')
```

```python
# logarithmic transformation
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
temp = img_gray.copy()

height = img_gray.shape[0]
width = img_gray.shape[1]

for i in range(0, height):
    for j in range(0, width):
        img_gray[i][j] = np.log(1+img_gray[i][j])

plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(temp, cmap='gray')
plt.title('original image')
plt.subplot(1, 2, 2)
plt.imshow(img_gray, cmap='gray')
plt.title('logarithmic transformation')
```

```python
# plot power-law plotting
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
r = np.unique(img_gray)

# get the values of s for r
s = np.zeros(len(r))
s_125 = np.zeros(len(r))
s_25 = np.zeros(len(r))
s_50 = np.zeros(len(r))
s25 = np.zeros(len(r))
s50 = np.zeros(len(r))
s10 = np.zeros(len(r))
for i in range(0, len(r)):
    s[i] = np.power(r[i], 1)
    s_125[i] = np.power(r[i], 0.125)
    s_25[i] = np.power(r[i], 0.25)
    s_50[i] = np.power(r[i], 0.5)
    s25[i] = np.power(r[i], 2.5)
    s50[i] = np.power(r[i], 5)
    s10[i] = np.power(r[i], 10)

plt.figure(figsize=(20, 20))
plt.subplot(2, 3, 1)
plt.plot(r, s_125)
plt.title('gamma=0.125')
plt.subplot(2, 3, 2)
plt.plot(r, s_25)
plt.title('gamma=0.25')
plt.subplot(2, 3, 3)
plt.plot(r, s_50)
plt.title('gamma=0.5')
plt.subplot(2, 3, 4)
plt.plot(r, s25)
plt.title('gamma=2.5')
plt.subplot(2, 3, 5)
plt.plot(r, s50)
plt.title('gamma=5')
plt.subplot(2, 3, 6)
plt.plot(r, s10)
plt.title('gamma=10')
```
