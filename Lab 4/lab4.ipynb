```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from numpy import asarray

#2D array using numpy with numpy 
array = np.array([[9,5,6,7,2],[1,1,0,1,2],[50,21,25,99,100],[150,255,250,130,200],[145,135,139,190,180]])
temp=np.array([[9,5,6,7,2],[1,1,0,1,2],[50,21,25,99,100],[150,255,250,130,200],[145,135,139,190,180]])
print(temp)


#avg of sum of array
avg = np.average(array)
#ceil the avg value
#avg = np.ceil(avg)
print(avg)
#thresholding the array with avg value
for i in range(0,5):
    for j in range(0,5):
        if array[i,j] < avg:
            array[i,j] = 0
        else:
            array[i,j] = 255
print(array)
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(temp)
plt.subplot(1,2,2)
plt.imshow(array)

```

```python
img=cv2.imread('/Users/mdarikrayhan/Downloads/Lab 3/Mona_Lisa.jpg')
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#average value of img_gray
avg = np.average(img_gray)

height, width = img_gray.shape
for i in range(0,height):
    for j in range(0,width):
        if img_gray[i,j] < avg:
            img_gray[i,j] = 0
        else:
            img_gray[i,j] = 255

plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(img)
plt.subplot(1,2,2)
plt.imshow(img_gray)
```
