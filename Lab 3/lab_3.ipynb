```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from numpy import asarray

img_cv2=cv2.imread('/Users/mdarikrayhan/Downloads/Lab 3/Mona_Lisa.jpg')
img_gray=cv2.cvtColor(img_cv2, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
```

```python
BGR_to_RGB=cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB)
image_flip = cv2.flip(BGR_to_RGB, 1)#step-2: flip image on y-axis
plt.imshow(image_flip)
```

```python
#step-3: Create another matrix with same dimentions
height=img_gray.shape[0]#no of rows
width=img_gray.shape[1]#no of colums
img_flip=np.zeros((height,width))

#step-4: flip img_gray on y-axis
for i in range(height):
    for j in range(width):
        img_flip[i][j]=img_gray[i][width-j-1]

plt.imshow(img_flip, cmap='gray')
```

```python
#step-5: show original,gray,fliped and manually flipped image
plt.figure(figsize=(20,20))
plt.subplot(2,2,1)
plt.imshow(BGR_to_RGB)
plt.title('Original Image')
plt.subplot(2,2,2)
plt.imshow(img_gray)
plt.title('Gray Image')
plt.subplot(2,2,3)
plt.imshow(image_flip)
plt.title('Flipped Image')
plt.subplot(2,2,4)
plt.imshow(img_flip )
plt.title('Manually Flipped Image')
plt.show()
```
