```python
# Title: CT03 - Unsharp Masking
import numpy as np
def round(avg):
    frac = avg - np.floor(avg)
    sign = np.sign(avg)
    if abs(frac) < 0.5:
        roundup = np.floor(abs(avg))
    else:
        roundup = np.ceil(abs(avg))
    return sign*roundup

image = np.array([[4,3,5],[3,6,2],[5,4,7]])
print("Original image:")
print(image)

print("\nKernel:")
kernel = np.array([[1,2,1],[2,4,2],[1,2,1]])
print(kernel,end="\n\n")


blurred = image.copy()
#pad the image with zeros
blurred = np.pad(blurred,1,mode='constant',constant_values=0)
for i in range(1,blurred.shape[0]-1):
    for j in range(1,blurred.shape[1]-1):
        temp = blurred[i-1:i+2,j-1:j+2]
        values = temp*kernel
        for e in range(values.shape[0]):
            for f in range(values.shape[1]):
                print(values[e,f],end=" ")
                if e == values.shape[0]-1 and f == values.shape[1]-1:
                    print("=",end=" ")
                else:
                    print("+",end=" ")
        sum = np.sum(values)
        print(sum,end=" ")
        sum = sum / 16
        print("=",sum,end=" ")
        rounded = round(sum)
        print("=",rounded,end="\n")
        blurred[i,j] = rounded
        
#remove the padding
blurred = blurred[1:-1,1:-1]
print("\nBlurred image:")
print(blurred)

#mask of image 
mask = image - blurred
print("\nMask:")
print(mask)

#output image with k=2.5
k = 2.5
output = np.zeros(image.shape)
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        output[i,j] = round(image[i,j] + k*mask[i,j])

print("\nOutput image:")
print(output)
```
