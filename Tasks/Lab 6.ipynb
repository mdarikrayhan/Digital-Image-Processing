```python
# Mean filter(Linear)
import numpy as np

# input 2D array
row = int(input("Enter the number of rows:"))
col = int(input("Enter the number of columns:"))
print("Enter the entries in a single line (separated by space): ")
entries = list(map(int, input().split()))
image = np.array(entries).reshape(row, col)

# padding
image = np.pad(image, 1, 'constant', constant_values=0)

# input kernel size
krow = int(input("Enter the number of rows of kernel:"))
kcol = int(input("Enter the number of columns of kernel:"))

# input the stride size
stride = int(input("Enter the stride size:"))

# output with same image
stepcount = 1
for i in range(1, row+1, stride):
    for j in range(1, col+1, stride):
        sum = np.sum(image[i-1:i+krow-1, j-1:j+kcol-1])
        avg = sum/(krow*kcol)
        
        frac = avg - np.floor(avg)
        if frac < 0.5:
            roundup = np.floor(avg)
        else:
            roundup = np.ceil(avg)
            
        print("Step", stepcount, ":", " Sum =", sum, "Average = ",avg," Round up-->", roundup)
        stepcount += 1
        image[i][j] = roundup


# remove padding
image = image[1:row+1, 1:col+1]
print(image)
```

```python
#weighted filter(Linear)
import numpy as np

# input 2D array
row = int(input("Enter the number of rows:"))
col = int(input("Enter the number of columns:"))
print("Enter the Image entries in a single line (separated by space): ")
entries = list(map(int, input().split()))
image = np.array(entries).reshape(row, col)

# padding
image = np.pad(image, 1, 'constant', constant_values=0)

# input kernel size
krow = int(input("Enter the number of rows of kernel:"))
kcol = int(input("Enter the number of columns of kernel:"))
print("Enter the Kernal entries in a single line (separated by space): ")
entries = list(map(int, input().split()))
weight = np.array(entries).reshape(krow, kcol)
div = np.sum(weight)


# input the stride size
stride = int(input("Enter the stride size:"))


# output with same image
stepcount = 1
for i in range(1, row+1, stride):
    for j in range(1, col+1, stride):
        sum = np.sum(image[i-1:i+krow-1, j-1:j+kcol-1]*weight)
        avg = sum/div
        
        frac = avg - np.floor(avg)
        if frac < 0.5:
            roundup = np.floor(avg)
        else:
            roundup = np.ceil(avg)
            
        print("Step", stepcount, ":", " Sum =", sum, "Average = ",avg," Round up-->", roundup)
        stepcount += 1
        image[i][j] = roundup

# remove padding
image = image[1:row+1, 1:col+1]
print(image)
```

```python
#Median filter(Non-Linear)
import numpy as np

# input 2D array
row = int(input("Enter the number of rows:"))
col = int(input("Enter the number of columns:"))
print("Enter the entries in a single line (separated by space): ")
entries = list(map(int, input().split()))
image = np.array(entries).reshape(row, col)

# padding
image = np.pad(image, 1, 'constant', constant_values=0)

# input kernel size
krow = int(input("Enter the number of rows of kernel:"))
kcol = int(input("Enter the number of columns of kernel:"))

# input the stride size
stride = int(input("Enter the stride size:"))

# output with same image
stepcount = 1
for i in range(1, row+1, stride):
    for j in range(1, col+1, stride):
        median = np.median(image[i-1:i+krow-1, j-1:j+kcol-1])
        print("Step", stepcount, ":", " Median =", median)
        stepcount += 1
        image[i][j] = median

# remove padding
image = image[1:row+1, 1:col+1]
print(image)
```
