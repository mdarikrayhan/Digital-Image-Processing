```python
import numpy as np

array = np.array([[1,8,8,0,7],[4,8,5,4,7],[5,0,6,2,6],[4,1,3,7,5],[0,1,0,2,0]])
array = np.pad(array, ((0,1),(0,1)), 'constant', constant_values=0)

for i in range(0,5):
    for j in range(0,5):
        print("for (",i,",",j,") :")
        gx = array[i][j+1] - array[i][j]
        gy = array[i+1][j] - array[i][j]
        print("gx = ",array[i][j+1],"-",array[i][j],"=",gx)
        print("gy = ",array[i+1][j],"-",array[i][j],"=",gy)
        
        magnitude = np.sqrt(gx**2 + gy**2)
        magnituderounded = np.round(magnitude)
        print("magnitude = ",magnitude,"=",magnituderounded)
        theta = np.arctan2(gy,gx)
        angle = theta * 180 / np.pi
        angle = np.round(angle,2)
        print("angle = tan^-1(",gy,"/",gx,") = ",angle,"degrees")

        print("")
```
