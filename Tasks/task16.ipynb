```python
import numpy as np

original = np.array([[50,60,20,200],[6,6,225,112],[237,50,100,90],[50,130,100,150]])

#gaussian filter
gaussian_filter = np.array([[1,2,1],[2,4,2],[1,2,1]])/16


blurred = original.copy()
blurred = np.pad(blurred,((1,1),(1,1)),'constant',constant_values=0)
for i in range(1,blurred.shape[0]-1):
    for j in range(1,blurred.shape[1]-1):
        #show the calculation
        blurred[i,j] = np.sum(blurred[i-1:i+2,j-1:j+2]*gaussian_filter)
blurred = blurred[1:-1,1:-1]

#calculate the mask
mask = original - blurred

#calculate the sharpened image
k = 4
sharpened = original + k*mask

print("Original image:")
print(original)
print("Blurred image:")
print(blurred)
print("Mask:")
print(mask)
print("Sharpened image:(K=4)")
print(sharpened)

```
