```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('/Users/mdarikrayhan/Documents/GitHub/Digital-Image-Processing/Tasks/panda.jpg')

array = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#array = np.array([[50,50,100,100],[50,50,100,100],[50,50,100,100],[50,50,100,100]])

#Roberts operator
roberts = np.array([[-1,0],[0,1]])

#Sobel operator
sobel = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])

#Prewitt operator
prewitt = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])


#calculate for roberts out image
roberts_out = array.copy()
roberts_out = np.pad(roberts_out,((0,1),(0,1)),'constant')
for i in range(roberts_out.shape[0]):
    for j in range(roberts_out.shape[1]):
        roberts_out[i,j] = np.sum(roberts*roberts_out[i:i+2,j:j+2])
roberts_out = roberts_out[:-1,:-1]
print("Roberts operator output:")
plt.imshow(roberts_out,cmap='gray')
print()
```

```python
#calculate for sobel out image
sobel_out = array.copy()
sobel_out = np.pad(sobel_out,((1,1),(1,1)),'constant')
for i in range(sobel_out.shape[0]-2):
    for j in range(sobel_out.shape[1]-2):
        sobel_out[i,j] = np.sum(sobel*sobel_out[i:i+3,j:j+3])

sobel_out = sobel_out[1:-1,1:-1]
print("Sobel operator output:")
plt.imshow(sobel_out,cmap='gray')        
```

```python
#calculate for prewitt out image
prewitt_out = array.copy()
prewitt_out = np.pad(prewitt_out,((1,1),(1,1)),'constant')
for i in range(prewitt_out.shape[0]-2):
    for j in range(prewitt_out.shape[1]-2):
        prewitt_out[i,j] = np.sum(prewitt*prewitt_out[i:i+3,j:j+3])

prewitt_out = prewitt_out[1:-1,1:-1]
print("Prewitt operator output:")
plt.imshow(prewitt_out,cmap='gray')        
```
